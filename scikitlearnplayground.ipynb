```python
# convert string into numerical values to fit our machine learning algo
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categories = ["Make", "Colour", "Doors"] # the data we are trying to convert into numerical values
one_hot = OneHotEncoder()
transform = ColumnTransformer([("one_hot", one_hot, categories)], remainder="passthrough")
transform_x = transform.fit_transform(X)
transform_x
```

```python
np.random.seed(42) #So we can use the same data again and use the transform_x to train our ML algo
X_train, X_test, y_train, y_test = train_test_split(transform_x, y, test_size=0.2)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
