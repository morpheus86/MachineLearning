```python
car_sales = pd.read_csv("./exported-car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,]|\.\d*', '').astype(int)
car_sales
```

```python
del car_sales["price"]
```

```python
car_sales
```

```python
car_sales["Price"].plot()
```

```python
car_sales["Total Sales"] = np.cumsum(car_sales["Price"])
car_sales
```

```python
car_sales["Seales Dates"] = pd.date_range("1/1/2020",periods=len(car_sales))
car_sales
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(car_sales["Seales Dates"], car_sales["Total Sales"])
```
