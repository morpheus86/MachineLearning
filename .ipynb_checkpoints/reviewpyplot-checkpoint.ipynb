```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot()
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.random.randint(1,5, size=(5))
y = np.random.randn(5)
plt.plot(x, y);
```

```python
fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
fig, ax = plt.subplots()
ax.set(title="simple array", xlabel="num", ylabel="numlessthan1")
ax.plot(x, y)
```

```python
fig.savefig("randomgraph.png")
```

```python
even_arr = np.arange(0,100,2)
even_arr
```

```python
fig, ax = plt.subplots()
ax.plot(even_arr, even_arr^2);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(even_arr, np.exp(even_arr));
```

```python
fig, ax = plt.subplots()
ax.scatter(even_arr, np.sin(even_arr))
```

```python
fav_foods = {
    "bananas": 0.4,
    "ice-cream": 5,
    "grilled-chicken": 12
}
fig, ax = plt.subplots(figsize=(10, 6))
ax.set(title="favorite foods", xlabel="foods", ylabel="prices")
ax.bar(fav_foods.keys(), fav_foods.values(), align="horizontal");
```
