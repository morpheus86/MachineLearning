##MATPLOTLIB EXAMPLE WORK FLOW

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1 Prepare data
x=[2,3,5,7]
y=[11,22,33,44]

# 2 setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3 plot data
ax.plot(x,y)

# 4 customize plot
ax.set(title="Simple Plot",
      xlabel="X-axis",
      ylabel="Y-axis")

# 5 save and show 
fig.savefig("original/simple-plot.png")
```

##MAKING figures with numpy array

```python
x = np.linspace(0,10,100)
```

```python
# Line Plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# SCatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make plot from a dictionary
nut_butter_prices = {
    "Almond butter": 10,
    "Peanut butter": 8,
    "Cashew butter": 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter store", ylabel="Price ($)");
```

```python
# Histogram plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplots options
# first option
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.hist(np.random.randn(1000))
ax4.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
# 2nd option
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].hist(np.random.randn(1000))
ax[1,1].bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

## HEART DISEASE DATA SET

```python
heart_disease = pd.read_csv("./original copy.csv")
heart_disease.to_csv("heart_disease.csv", index=False)
```

```python
heart_disease = pd.read_csv("./heart_disease.csv")
heart_disease.head()
```

## Create a histogram to see distribution of age col

```python
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
heart_disease.head(10)
```

```python
sample = heart_disease.sample(frac=0.3, random_state=2)
sample
```

```python
sample.plot.hist(figsize=(20, 20),bins=20);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Pyplot metod
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# OO method with pyplot
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
ax.set_xlim([45,100])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))
# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
#  customize plot
ax.set(title="Heart Disease and Chlesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")
# Add Legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizotal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(15,15), sharex=True)

# Add data
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize ax0
ax0.set(title="Heart disease and cholesteorl Levels",
       xlabel="Age",
       ylabel="cholesterol")
# Add legend
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a dotted line
ax0.axhline(over_50["chol"].mean(), linestyle="--")

scatter2 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax1.set(title="Heart disease and thalach Levels", xlabel="age", ylabel="Max heart rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle="-.")

# Add Title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
